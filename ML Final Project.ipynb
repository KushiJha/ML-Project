```python
import pandas as pd
import os
```

```python
external_path = os.path.join('External_Cibil_Dataset.xlsx')
internal_path = os.path.join('Internal_Bank_Dataset.xlsx')
unseen_path = os.path.join('Unseen_Dataset.xlsx')
```

```python
external_df = pd.read_excel(external_path)
internal_df = pd.read_excel(internal_path)
unseen_df = pd.read_excel(unseen_path)
```

```python
# STEP 3: Merge External and Internal Datasets on 'PROSPECTID'
combined_df = pd.merge(external_df, internal_df, on='PROSPECTID', how='inner')
```

```python
combined_df
```

```python
combined_df.isnull().sum()
```

```python
# STEP 4: Handle missing values (-99999 and NaNs)
combined_df.replace(-99999, pd.NA, inplace=True)
combined_df.dropna(axis=1, thresh=0.8 * len(combined_df), inplace=True)
combined_df.fillna(combined_df.median(numeric_only=True), inplace=True)
```

```python
combined_df
```

```python
sample_input = combined_df[common_cols].iloc[0].to_dict()
print("✅ Sample input:\n", sample_input)
```

```python
label_encoders = {}
categorical_cols = combined_df.select_dtypes(include='object').columns.drop('Approved_Flag', errors='ignore')
```

```python
# STEP 5: Encode categorical features
from sklearn.preprocessing import LabelEncoder
```

```python
categorical_cols
```

```python
for col in categorical_cols:
    le = LabelEncoder()
    combined_df[col] = le.fit_transform(combined_df[col].astype(str))
    label_encoders[col] = le
```

```python
le
```

```python
# STEP 6: Align features with unseen dataset
common_cols = [col for col in combined_df.columns if col in unseen_df.columns and col not in ['PROSPECTID', 'Approved_Flag']]
X = combined_df[common_cols]
```

```python
X
```

```python
# Encode Target Column
target_le = LabelEncoder()
y = combined_df['Approved_Flag'].astype(str).str.strip()
y_encoded = target_le.fit_transform(y)
print("✅ Target classes in LabelEncoder:", target_le.classes_)
```

```python
# STEP 7: Train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
X_train
y_train
```

```python
# STEP 8: Scale features for Logistic Regression
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
# STEP 9: Train Multiple Models
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=2000),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "SVM": SVC(kernel='linear', max_iter=1000)
}
```

```python
for name, model in models.items():
    print(f"\n{name} Results:")

    if name == "Logistic Regression":
        model.fit(X_train_scaled, y_train)
        preds = model.predict(X_test_scaled)
    else:
        model.fit(X_train, y_train)
        preds = model.predict(X_test)

    print(classification_report(y_test, preds, target_names=[str(cls) for cls in target_le.classes_]))
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Define models
models = {
    "Logistic Regression": LogisticRegression(max_iter=2000),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "SVM": SVC(kernel='linear', max_iter=1000)
}

# Train and evaluate each model
for name, model in models.items():
    print(f"\n{name} Accuracy:" )

    
    # Logistic Regression requires scaled features
    if name == "Logistic Regression":
        model.fit(X_train_scaled, y_train)
        preds = model.predict(X_test_scaled)
    else:
        model.fit(X_train, y_train)
        preds = model.predict(X_test)
    
    accuracy = accuracy_score(y_test, preds)
    print(f"Accuracy: {accuracy * 100:.2f}%")

```

```python
best_model = models["Random Forest"]
```

```python
from joblib import dump

# Save the best trained model
dump(best_model, 'best_model.joblib')

# Save the label encoders used for categorical columns
dump(label_encoders, 'label_encoders.joblib')

# Save the list of common feature columns used in training
dump(common_cols, 'common_cols.joblib')

# Save the label encoder used for the target variable
dump(target_le, 'target_le.joblib')
```

```python
import gradio as gr
import pandas as pd
from joblib import load

# Step 2.1: Load saved files
best_model = load('best_model.joblib')
label_encoders = load('label_encoders.joblib')
common_cols = load('common_cols.joblib')
target_le = load('target_le.joblib')

# Step 2.2: Prediction function
def predict_risk(**kwargs):
    df = pd.DataFrame([kwargs])

    # Apply encoders to categorical columns
    for col, le in label_encoders.items():
        if col in df.columns:
            df[col] = df[col].astype(str)
            df[col] = df[col].apply(lambda x: x if x in le.classes_ else le.classes_[0])
            df[col] = le.transform(df[col])
    
    # Keep only required columns
    df = df[common_cols]

    # Predict and decode label
    pred = best_model.predict(df)
    label = target_le.inverse_transform(pred)[0]

    # Human-readable risk category
    risk_mapping = {
        "P1": "P1 — High Risk: Likely to default, consider strong caution.",
        "P2": "P2 — Moderate Risk: Good candidate, but with minor potential issues.",
        "P3": "P3 — Low Risk: Generally safe with good credit behavior.",
        "P4": "P4 — Very Low Risk: Excellent credit profile, minimal risk."
    }
    return risk_mapping.get(label, f"Prediction: {label}")

# Step 2.3: Create dynamic input UI
inputs = []
input_keys = []

for col in common_cols:
    if col in label_encoders:
        inputs.append(gr.Dropdown(choices=label_encoders[col].classes_.tolist(), label=col))
    else:
        inputs.append(gr.Number(label=col))
    input_keys.append(col)

# Step 2.4: Wrapper for unpacking Gradio input
def gradio_wrapper(*args):
    return predict_risk(**dict(zip(input_keys, args)))

# Step 2.5: Build and launch interface
demo = gr.Interface(
    fn=gradio_wrapper,
    inputs=inputs,
    outputs=gr.Textbox(label="Predicted Risk Category"),
    title="Credit Risk Prediction App",
    description="Enter the applicant's details to predict the credit approval category (P1–P4)."
)

demo.launch()
```
